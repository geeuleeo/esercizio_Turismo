```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt;

dArr = pd.read_csv('https://raw.githubusercontent.com/FabioGagliardiIts/datasets/main/dati_turismo/Arrivi-negli-agriturismi-in-Italia-per-regione.csv', sep=";")

stats = dArr.groupby('Regione')['Arrivi'].agg(['mean', 'std', 'median'])

# Stampa le statistiche
print(stats)

# Visualizza le statistiche in un grafico
stats.plot(kind='bar', subplots=True, layout=(3,1), figsize=(10,10))
plt.tight_layout()
plt.show()
```
