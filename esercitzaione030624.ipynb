```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
,
# Nome del file CSV per una singola struttura (Agriturismi)
file_name = '/content/Arrivi-nei-campeggi-e-villaggi-turistici-in-italia-per-regione.csv'

# Leggi i dati dal file CSV
data = pd.read_csv(file_name, delimiter=';')

# Filtrare le colonne di interesse (Anno, Regione, Presenze)
data = data[['Anno', 'Regione', 'Arrivi']]

# Raggruppare per 'Regione' e 'Anno' e sommare le 'Presenze'
data_grouped = data.groupby(['Regione', 'Anno']).sum().reset_index()

# Pivot per creare una matrice dove le righe sono le regioni e le colonne sono gli anni
pivot_df = data_grouped.pivot(index='Regione', columns='Anno', values='Arrivi')

# Riempire eventuali valori NaN con 0 o con un valore medio
pivot_df = pivot_df.fillna(0)

# Convertire il DataFrame in una matrice NumPy
data_matrix = pivot_df.to_numpy()

# Calcolare la matrice di correlazione
corr_matrix = np.corrcoef(data_matrix)

# Stampare la matrice di correlazione
#print(corr_matrix)

# Visualizzare la matrice di correlazione con una heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', xticklabels=pivot_df.index, yticklabels=pivot_df.index)
plt.title('Matrice di Correlazione delle Presenze Turistiche per Regione (Agriturismi)')
plt.show()
```
